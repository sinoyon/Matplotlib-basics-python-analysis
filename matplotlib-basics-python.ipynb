# Matplotlib basics python for beginners

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
#plt.show()
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
x,y
```

```python
plt.plot(x,y, 'r')
plt.xlabel('x axis')
plt.ylabel('y axix')
plt.title('plot Title is here')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.subplot(1,2,2)
plt.plot(y,x)
```

```python
from matplotlib import axes
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,'r')
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.set_title('Title of the plot')
plt.show()
```

```python
fig=plt.figure()

axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5, 0.4,0.3])

axes1.plot(x,y,'r')
axes1.set_xlabel('X axis')
axes1.set_ylabel('Y axis')
axes1.set_title('Title of graph')

axes2.plot(x,y,'g')
axes2.set_xlabel('X axis')
axes2.set_ylabel('Y axis')
axes2.set_title('Title of graph in')
plt.show()
```

```python
fig, axes=plt.subplots()

axes.plot(x,y,'r')
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.set_title('Title of the plot')
```

```python
fig,axes=plt.subplots(nrows=1, ncols=2)
```

```python
axes
```

```python
for ax in axes:
  ax.plot(x,y,'r')
  ax.set_xlabel('x_lable')
  ax.set_ylabel('y_lable')
  ax.set_title('this is the plot title')
```

```python
fig=plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label="X Squared")
ax.plot(x, x**3, label="x cubed") 
ax.legend()
```

```python
fig,ax=plt.subplots()
ax.plot(x,x**2,'r.-',label="X Squared")
ax.plot(x,x**3,'b--', label="x cubed")
ax.legend()
```

```python
fig, ax=plt.subplots()
ax.plot(x+2, color="#FF8C00")
ax.plot(x+3)
ax.plot(x+4)
plt.show()
```

```python
#from matplotlib.markers import MarkerStyle
fig, ax=plt.subplots(figsize=(12,6))
ax.plot(x,x+1, color="red", lw=1.00, ls='-.', marker="+", markersize=30, markerfacecolor="black",markeredgewidth=3,markeredgecolor="green")
plt.show()
```

```python
fig,axes=plt.subplots(1,3,figsize=(12,4))
axes[0].plot(x, x**2, x**3)
axes[0].set_title("Default range")

axes[1].plot(x, x**2, x**3)
axes[1].axis('tight')
axes[1].set_title("Tight axes")

axes[2].plot(x, x**2, x**3)
axes[2].set_ylim([0,60])
axes[2].set_xlim([2,5])
axes[2].set_title("Custom range")

plt.show()
```
